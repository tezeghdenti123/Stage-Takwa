```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity


df = pd.read_csv('job_listings.csv')


candidate_profile = """
Name: ALI ALI
Email: ALI.ALI@gmail.com
Phone:+216 55 555 555

Professional Summary:
Software engineer with over 5 years of experience in Python programming, machine learning, and web development. 
Proficient in frameworks like Django and Flask, with a solid understanding of SQL and NoSQL databases. 
Experienced in building scalable applications and working in agile teams. 
Strong problem-solving skills and a passion for learning new technologies.

Skills:
- Python
- Machine Learning
- Web Development
- Django
- Flask
- SQL
- NoSQL

Education:
- BSc in Computer Science
- MSc in Data Science
"""


df['job_info'] = df['Job Title'] + " " + df['Job Description']


vectorizer = TfidfVectorizer(stop_words='english')


tfidf_matrix = vectorizer.fit_transform(df['job_info'].tolist() + [candidate_profile])


cosine_similarities = cosine_similarity(tfidf_matrix[-1], tfidf_matrix[:-1]).flatten()


df['similarity_score'] = cosine_similarities


sorted_df = df.sort_values(by='similarity_score', ascending=False)


pd.set_option('display.max_rows', None)


print(sorted_df[['Company', 'Job Title', 'similarity_score']])
```
